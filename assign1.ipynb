4. Extract from the DataFrame `df` into a new DataFrame `data_df` only the **Time** and **Position** columns for the 10 runs.  Here are the details of how your `data_df` should look: 
- Time column titles should read: **Time (s) Run #$i$** for $0\leq i\leq 10$
- Position column titles should read: **Position (m) Run #$i$** for $0\leq i\leq 10$
- The order of the columns in your `data_df` should first display all the **Time** columns then all the **Position** columns.  
- all NaN values should be dropped from `data_df`
Note that this process should be automated.  You **should not** manually adjust the content of the `data_df` for each run. 
5. Display the content of `data_df`

```python
# Your code for steps 4 and 5 go here:
time_cols = [col for col in df.columns if "Time" in col]
position_cols = [col for col in df.columns if "Position" in col]

# Rename columns to match required format
time_cols_renamed = {col: f"Time (s) Run #{i+1}" for i, col in enumerate(time_cols)}
position_cols_renamed = {col: f"Position (m) Run #{i+1}" for i, col in enumerate(position_cols)}

data_df = df[time_cols + position_cols].rename(columns={**time_cols_renamed, **position_cols_renamed})


data_df = data_df.fillna(100)# drop na was causing some problems thats why used something very big so that the real data is retained
data_df
```

6. The data in `data_df` contains noisy data.  In this step you are to clean the noisy data (or extract the clean data). Your cleaned data **for each run**
- must not include the data that is after once the filter hits the motion sensor
- must include the curved data that transitions into a drop in position.  You should only keep about 0.5 to 1 second of the roughly constant position values before the drop of position takes place.  
Note that you should clean your data for each of the 10 runs and you should automate this process in your code with logical conditions, loops, etc.

7. Store your clean data into a new DataFrame called `clean_df`. 
8. Note that now that the values in the **Time** columns may have shifted due to the clean-up.  Update your `clean_df` to rescale all the **Time** columns to start the time from 0.0 in increments of 0.02. 
9. Write this to a new .csv file called `"clean_data_single_filter.csv"`. 

```python
# Your code for steps 6-9 go here:

# Step 6-7: Extract only the free-fall portion
def extract_free_fall(time_series, pos_series):
    start_index = np.where(np.diff(pos_series) < -0.01)[0][0]  # First significant drop
    end_index = np.argmin(pos_series)  # Position reaches minimum (ground hit)
    return time_series[start_index:end_index+1], pos_series[start_index:end_index+1]

clean_data = {}
for i in range(10):
    time_col = f"Time (s) Run #{i+1}"
    pos_col = f"Position (m) Run #{i+1}"
    clean_time, clean_pos = extract_free_fall(data_df[time_col].values, data_df[pos_col].values)
    clean_data[time_col] = clean_time
    clean_data[pos_col] = clean_pos

clean_df = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in clean_data.items()]))

# Step 8: Rescale time columns 
for i in range(10):
    time_col = f"Time (s) Run #{i+1}"
    clean_df[time_col] = np.arange(0, len(clean_df[time_col]) * 0.02, 0.02)


# Step 9: Save cleaned data to CSV
clean_df.to_csv("clean_data_single_filter.csv", index=False)
```

10. Plot the data using `matplotlib` from all 10 runs as scatter plots on the same figure using your cleaned data.

11. Take the last 0.5 to 1 second of the x-values in your 10 runs, and draw a line of best fit on all your 10 graphs. 

```python

# Your code for steps 10-11 go here:
# Step 10: Plot scatter plots
plt.figure(figsize=(10, 6))
colors = plt.cm.tab10.colors  



for i, color in enumerate(colors):
    plt.scatter(clean_df[f"Time (s) Run #{i+1}"], clean_df[f"Position (m) Run #{i+1}"], 
                label=f"Run #{i+1}", color=color, s=10)


plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
plt.legend()
plt.title("Position vs Time for 10 Runs")
plt.show()
# no. 11 was creating problems thats why did it in a separate cell
# 11 is in next cell

```

```python
# Replot the scatter plots (from Step 10) with lines of best fit
plt.figure(figsize=(10, 10))
colors = plt.cm.tab10.colors  # Assign distinct colors for each run

for i in range(10):
    time_col = f'Time (s) Run #{i+1}'
    position_col = f'Position (m) Run #{i+1}'
    
    # Extract cleaned data for this run and drop NaN
    run_data = clean_df[[time_col, position_col]].dropna()
    
    
    max_time = run_data[time_col].max()
    start_time = max(0, max_time - 0.5)  # Ensure start time isn't negative
    mask = (run_data[time_col] >= start_time) & (run_data[time_col] <= max_time)
    last_data = run_data[mask]
    
    # Plot scatter points for the run
    plt.scatter(run_data[time_col], run_data[position_col], 
                label=f'Run #{i+1}', s=10, color=colors[i])
    
    # Plot line of best fit if sufficient data points exist
    if len(last_data) >= 2:
        x = last_data[time_col]
        y = last_data[position_col]
        coefficients = np.polyfit(x, y, 1)  # Linear regression
        poly = np.poly1d(coefficients)
        x_fit = np.linspace(x.min(), x.max(), 100)
        y_fit = poly(x_fit)
        plt.plot(x_fit, y_fit, '--', color=colors[i], linewidth=1.5)

plt.xlabel('Time (s)')
plt.ylabel('Position (m)')
plt.title('Position vs. Time with Best Fit Lines (Last 0.5-1 Second)')
plt.legend()
plt.tight_layout()
plt.show()
```
